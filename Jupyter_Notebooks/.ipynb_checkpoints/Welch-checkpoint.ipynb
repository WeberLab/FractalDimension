```python
example_slice = '/home/weberam2/Dropbox/AssistantProf_BCCHRI/Projects/Fractal/FractalData/2020-09-17_BCCHR_test_AlexW/derivatives/func/FracTool_run1/filtered_func_data_clean.nii.gz'
slice_img = nib.load(example_slice, mmap=False)
slice_array = slice_img.get_fdata()
slice_sq = slice_array[20:40,10:20,30,:]
```

```python
voxel = slice_array[30,13,30,:]
```

```python
plt.plot(voxel)
```

```python
print(slice_img.header.get_zooms())
```

```python
TR = slice_img.header.get_zooms()
TR = TR[3]
print(TR)
```

```python
nperseg = math.floor(len(voxel)/8)
print(nperseg)
```

```python
noverlap = math.floor(nperseg/2)
print(noverlap)
```

```python
math.floor(len(voxel))
```

```python
w = welch(voxel, fs = TR, nperseg = nperseg, noverlap = noverlap)
```

```python
plt.loglog(w[0],w[1])
```

```python
from sklearn.linear_model import LinearRegression
```

```python
x = np.log10(w[0])[1:].reshape((-1, 1))
y = np.log10(w[1])[1:]
print(x)
print(y)
```

```python
model = LinearRegression().fit(x, y)
```

```python
model.coef_
```

```python
((model.coef_)*(-1) - 1)/2
```

```python
y_new = model.predict(x)
```

```python
# plot the results
plt.figure(figsize=(4, 3))
ax = plt.axes()
ax.scatter(x, y)
ax.plot(x, y_new)

ax.set_xlabel('x')
ax.set_ylabel('y')

ax.axis('tight')


plt.show()
```
